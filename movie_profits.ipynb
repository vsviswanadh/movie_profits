```python
df.groupby("genre")["production_budget"].sum().plot(kind="bar")
plt.title("sum of production_budget of each genre")
plt.xlabel("genre")
plt.ylabel("production budget")
plt.xticks(rotation=0)
plt.show()
```

```python
df.groupby(df["release_date"].str.slice(6,8))["worldwide_gross"].mean().plot(kind="line",color="red")
plt.title("average gross per year")
plt.xlabel("release year")
plt.ylabel("average gross")
plt.show()
```

```python
df.groupby("mpaa_rating")["domestic_gross"].sum().plot(kind="pie")
plt.title("total domestic gross")
plt.ylabel("")
plt.show()
```

```python
df.plot(kind="scatter",x="production_budget",y="worldwide_gross")
plt.show()
```

```python
df.groupby("distributor")["domestic_gross"].mean().head().plot(kind="barh")
plt.title("average domestic gross by distributor")
plt.xlabel("average domestic gross")
plt.ylabel("distributor")
plt.show()
```
